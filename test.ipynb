```python
import subprocess
from moviepy.editor import VideoFileClip, AudioFileClip

def replace_audio_with_moviepy(video_path, new_audio_path, output_path):
    # Load video clip
    video_clip = VideoFileClip(video_path)
    
    # Load new audio
    new_audio = AudioFileClip(new_audio_path)
    
    # Set the audio of the video clip
    video_with_new_audio = video_clip.set_audio(new_audio)
    
    # Write the result to a file
    video_with_new_audio.write_videofile(output_path, audio_codec='aac')

# Usage
replace_audio_with_moviepy("video.mp4", "audio.mp3", "output_moviepy.mp4")

# replace_audio_in_video("video.mp4", "audio.mp3", 'new_video.mp4')
```
